```python
import PyPDF2
import matplotlib.pyplot as plt
import numpy as np
import textstat
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
# Function to extract text from a PDF
def extract_text_from_pdf(pdf_path):
    with open(pdf_path, "rb") as file:
        pdf_reader = PyPDF2.PdfReader(file)
        text = "".join(page.extract_text() or "" for page in pdf_reader.pages)
    return text

# Function to extract top N keywords using TF-IDF
```

```python
def extract_keywords(text, top_n=10):
    vectorizer = TfidfVectorizer(stop_words="english", max_features=top_n)
    tfidf_matrix = vectorizer.fit_transform([text])
    keywords = vectorizer.get_feature_names_out()
    return keywords

# Function to compute TF-IDF similarity
```

```python
def compute_similarity(job_desc_text, resume_text):
    vectorizer = TfidfVectorizer(stop_words="english")
    tfidf_matrix = vectorizer.fit_transform([job_desc_text, resume_text])
    similarity = cosine_similarity(tfidf_matrix[0], tfidf_matrix[1])[0][0]
    return similarity * 100  # Convert to percentage

# Function to calculate ATS score (Readability Score)
```

```python
def compute_ats_score(text):
    return textstat.flesch_reading_ease(text)

# Function to visualize similarity and ATS scores
def plot_similarity(resume_scores, resume_ats_scores):
    plt.figure(figsize=(8, 5))
    
    resumes = list(resume_scores.keys())
    similarity_scores = list(resume_scores.values())
    ats_scores = list(resume_ats_scores.values())

    # X locations for groups of bars
    x = np.arange(len(resumes))

    # Bar width
    width = 0.4

    # Plot similarity and ATS scores
    plt.bar(x - width/2, similarity_scores, width, label="Similarity Score (%)", color="skyblue")
    plt.bar(x + width/2, ats_scores, width, label="ATS Score", color="orange")

    # Labels and title
    plt.xticks(x, resumes, rotation=45, ha="right")
    plt.ylim(0, 100)
    plt.ylabel("Score (%)")
    plt.title("Resume Similarity & ATS Readability Score")

    # Display values on bars
    for i, (sim_score, ats_score) in enumerate(zip(similarity_scores, ats_scores)):
        plt.text(i - width/2, sim_score + 2, f"{sim_score:.2f}%", ha='center', fontsize=10, fontweight='bold')
        plt.text(i + width/2, ats_score + 2, f"{ats_score:.2f}", ha='center', fontsize=10, fontweight='bold')

    plt.legend()
    plt.tight_layout()
    plt.show()

```

```python
# File Paths (Modify as per your file locations)
job_desc_path = "JD2.pdf"
resume_paths = ["Res2.pdf", "SAM_T_JAMES_Resume_Updated_March - Copy.pdf"]  # Add multiple resumes

# Extract and process text
job_description_text = extract_text_from_pdf(job_desc_path)
resume_texts = {res: extract_text_from_pdf(res) for res in resume_paths}
```

```python
# Print extracted text (trimmed for readability)
print("\n🔹 Extracted Job Description (First 500 chars):\n", job_description_text[:500], "\n...")
for res, text in resume_texts.items():
    print(f"\n🔹 Extracted Resume ({res}) (First 500 chars):\n", text[:500], "\n...")

# Extract key phrases from the job description and resumes
job_keywords = extract_keywords(job_description_text, top_n=10)
print("\n📌 Key Job Description Keywords:", job_keywords)

resume_keywords = {res: extract_keywords(text, top_n=10) for res, text in resume_texts.items()}
for res, keywords in resume_keywords.items():
    print(f"📌 Key Resume Keywords ({res}):", keywords)

# Compute similarity for each resume
resume_scores = {res: compute_similarity(job_description_text, text) for res, text in resume_texts.items()}

# Print and visualize results
for res, score in resume_scores.items():
    print(f"\n📊 Resume Similarity Score ({res}): {score:.2f}%")

plot_similarity(resume_scores)
    
# Compute ATS Readability Score for each resume
resume_ats_scores = {res: compute_ats_score(text) for res, text in resume_texts.items()}

# Print ATS Scores
for res, ats_score in resume_ats_scores.items():
    print(f"\n📑 ATS Readability Score ({res}): {ats_score:.2f}")

# Plot both similarity and ATS scores
plot_similarity(resume_scores, resume_ats_scores)

```
